```python
# ============================================================
# NOTEBOOK 5 — Final Visualization Dashboard (Side-by-Side Layout)
# ============================================================

from IPython.display import display, Markdown, Image, HTML
import pandas as pd
import os

# ---------- Setup ----------
output_folder = "./outputs/"

display(Markdown("# Text, Social Media & Web Analytics — Final Dashboard"))
```

```python
display(Markdown("### Consolidated Insights from Tweets, Reviews, and News Data\n"
"---\nThis dashboard summarizes sentiment, topic modeling, and cross-platform trends in a compact visual layout."))

# ---------- Utility Function ----------
def display_side_by_side(image_paths, titles=None, width=420):
    """Display multiple images side-by-side."""
    html = "<div style='display:flex; flex-wrap:wrap; justify-content:center; gap:20px;'>"
    for i, path in enumerate(image_paths):
        if os.path.exists(path):
            title_html = f"<div style='text-align:center; font-weight:bold; margin-top:8px;'>{titles[i] if titles else ''}</div>"
            html += f"""
                <div style='text-align:center;'>
                    <img src='{path}' width='{width}' style='border-radius:10px; box-shadow:0 2px 6px rgba(0,0,0,0.2); margin-bottom:5px;'/>
                    {title_html}
                </div>
            """
    html += "</div>"
    display(HTML(html))

# ============================================================
# Section 1: Sentiment Overview
# ============================================================

display(Markdown("## 1. Sentiment Analysis Overview"))

for name in ["tweets", "reviews", "news"]:
    path = os.path.join(output_folder, f"{name}_with_sentiment.csv")
    if os.path.exists(path):
        df = pd.read_csv(path)
        avg_sent = df["sentiment_score"].mean()
        pos_pct = (df["sentiment_label"] == "Positive").mean() * 100
        neg_pct = (df["sentiment_label"] == "Negative").mean() * 100
        display(Markdown(f"**{name.capitalize()}** — Avg Sentiment: {avg_sent:.2f} | Positive: {pos_pct:.1f}% | Negative: {neg_pct:.1f}%"))

sent_img = [os.path.join(output_folder, "sentiment_distribution.png"),
            os.path.join(output_folder, "sentiment_distribution_comparison.png")]
display_side_by_side(sent_img, ["Sentiment Spread", "Cross-Platform Sentiment Comparison"])

# ============================================================
# Section 2: Topic Modeling & Keyword Highlights
# ============================================================

display(Markdown("## 2. Topic Modeling & Keyword Highlights"))

topic_summary_path = os.path.join(output_folder, "topic_summary.csv")
if os.path.exists(topic_summary_path):
    topics_df = pd.read_csv(topic_summary_path)
    display(Markdown("### Top Keywords per Topic"))
    display(topics_df.head(10))

# Show keyword bar charts in a row
keyword_imgs = [
    os.path.join(output_folder, "tweets_top_keywords.png"),
    os.path.join(output_folder, "reviews_top_keywords.png"),
    os.path.join(output_folder, "news_top_keywords.png")
]
display_side_by_side(keyword_imgs, ["Tweets", "Reviews", "News"], width=440)

# ============================================================
# Section 3: Sentiment vs Topic Correlation
# ============================================================

display(Markdown("## 3. Sentiment vs Topic Correlation"))

topic_imgs = [
    os.path.join(output_folder, "tweets_sentiment_by_topic.png"),
    os.path.join(output_folder, "reviews_sentiment_by_topic.png"),
    os.path.join(output_folder, "news_sentiment_by_topic.png")
]
display_side_by_side(topic_imgs, ["Tweets", "Reviews", "News"], width=440)

heatmap_path = os.path.join(output_folder, "sentiment_topic_heatmap.png")
if os.path.exists(heatmap_path):
    display(Markdown("### Cross-Platform Sentiment-Topic Heatmap"))
    display(Image(filename=heatmap_path, width=700))

# ============================================================
# Section 4: Insights Summary
# ============================================================

display(Markdown("## 4. Analytical Summary"))

insights_path = os.path.join(output_folder, "sentiment_topic_insights.csv")
if os.path.exists(insights_path):
    insights_df = pd.read_csv(insights_path)
    display(insights_df)

display(Markdown("""
**Summary of Key Insights:**
- Sentiment varies widely across platforms, reflecting differing audience emotions.
- Reviews show the strongest emotional polarity, while news remains neutral.
- Top-performing topics align with positive sentiment clusters, revealing brand trust indicators.
- Visual comparison helps identify sentiment-driven topics that can shape marketing strategies.
"""))

# ============================================================
# Section 5: Combined WordCloud & Visual Gallery
# ============================================================

display(Markdown("## 5. WordCloud and Visual Gallery Overview"))

wordcloud_imgs = [
    os.path.join(output_folder, "tweets_wordcloud.png"),
    os.path.join(output_folder, "reviews_wordcloud.png"),
    os.path.join(output_folder, "news_wordcloud.png")
]
display_side_by_side(wordcloud_imgs, ["Tweets WordCloud", "Reviews WordCloud", "News WordCloud"], width=420)

display(Markdown("---"))
display(Markdown("### End of Dashboard — Text, Social Media & Web Analytics Project"))
display(Markdown("Prepared by: **Abhishek Gantana | MBA - Business Analytics | University of Hyderabad**"))
```
